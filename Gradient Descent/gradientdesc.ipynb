```python
import numpy as np
import matplotlib.pyplot as plt

def gradient_descent(x, y):
    w_curr = b_curr = 0
    rate = 0.01
    n = len(x)
    
    # Plot the initial data points
    plt.scatter(x, y, color='red', marker='+', linewidth=2, label='Data Points')
    
    for i in range(10000):
        y_predicted = w_curr * x + b_curr
        dw = -(2 / n) * sum(x * (y - y_predicted))
        db = -(2 / n) * sum(y - y_predicted)
        w_curr = w_curr - rate * dw
        b_curr = b_curr - rate * db

    # Plot the final regression line
    y_predicted = w_curr * x + b_curr
    plt.plot(x, y_predicted, color='green', label='Regression Line')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gradient Descent Linear Fit')
    plt.legend()
    plt.show()

# Test the function with numeric data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
gradient_descent(x, y)
```
