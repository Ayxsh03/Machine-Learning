```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
iris
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
X = df.iloc[:,:-1] # :-1 means all columns except the last one
Y = iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treemodel = DecisionTreeClassifier() # Post prunning
```

```python
treemodel.fit(X_train, Y_train)
```

```python
# Constructing Decision Tree
from sklearn import tree
plt.figure(figsize=(20,15))
tree.plot_tree(treemodel, filled=True, feature_names=df.columns, class_names=iris.target_names)
plt.show()
```

```python
# Hyperparameter tuning
treemod = DecisionTreeClassifier(max_depth=2)
treemod.fit(X_train, Y_train)
plt.figure(figsize=(10,5))
tree.plot_tree(treemod, filled=True, feature_names=df.columns, class_names=iris.target_names)
plt.show()
```

```python
y_pred = treemodel.predict(X_test)
y_pred2 = treemod.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score1 = accuracy_score(y_pred, Y_test)
score2 = accuracy_score(y_pred2, Y_test)
print('Accuracy1:', score1)
print('Accuracy2:', score2)
```

```python
print(classification_report(y_pred, Y_test))
print(classification_report(y_pred2, Y_test))
```
